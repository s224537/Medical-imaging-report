#### Window function 

##### WL 

```python
# Window-Level Function
def apply_window_level(image, window, level):
    lower_bound = level - (window / 2)
    upper_bound = level + (window / 2)
    # Clip the intensity values
    X_clipped = np.maximum(np.minimum(image, upper_bound), lower_bound)
    # Normalize the intensity values
    X_WL = (X_clipped - lower_bound) / (upper_bound - lower_bound)
    return X_WL
```

#### Scalebar

Function der indsÃ¦tter scalebar pÃ¥ billede

```python

def plot_with_scalebars(image, index, pixel_spacing, scalebars, plane):

    """
    Plots a slice image with multiple scalebars.
    
    Parameters:
    - image: 2D array-like
        The cropped slice to be plotted.
    - index: int
        Index of the slice.
    - pixel_spacing: tuple
        Pixel dimensions in (x, y) direction.
    - scalebars: list of dict or tuple of lists of dict
        List of scalebars, each defined as a dictionary with keys:
        'length_mm': float - Length of the scalebar.
        'position': tuple - (x_start, y_start) position of the scalebar.
        'direction': tuple - (x_direction, y_direction) direction vector for the scalebar.
    - plane: str
        Name of the plane (e.g., 'Axial', 'Coronal', 'Sagittal').
    """

    pixel_spacing_x = pixel_spacing[0]  # Pixel dimension in x-direction 
    pixel_spacing_y = pixel_spacing[1]  # Pixel dimension in y-direction

    # Handle scalebars input that might be a tuple of lists of dictionaries
    if isinstance(scalebars, tuple):
        scalebars = [item for sublist in scalebars for item in sublist]
    
    # Plotting the cropped slice with scalebars
    plt.figure(figsize=(10, 6))
    plt.imshow(image, cmap='gray')
    plt.title(f'{plane} Slice at Index {index} with scalebars')
    plt.xlabel('R-L [pixels]')  # Right to Left direction for x-axis
    plt.ylabel('A-P [pixels]')  # Anterior to Posterior direction for y-axis

    # Plot each scalebar
    for scalebar in scalebars:
        length = scalebar['length']
        (x_start, y_start) = scalebar['position']
        (x_direction, y_direction) = scalebar['direction']
        
        # Calculate scalebar length in pixels
        length_pixels_x = (length / pixel_spacing_x) * x_direction
        length_pixels_y = (length / pixel_spacing_y) * y_direction
        
        #print(length_pixels_x, length_pixels_y)

        x_end = x_start + length_pixels_x
        y_end = y_start + length_pixels_y

        # Plot the scalebar
        plt.plot([x_start, x_end], [y_start, y_end], color='red', linewidth=2)
        plt.text(x_start + 5, y_start - 5, f'{length}', color='white', fontsize=10, backgroundcolor='black')




```

```python


def visualize_slice(img_data, slice_index, plane):
    """
    Visualizes a slice from the provided image data based on the specified plane.
    
    Parameters:
    - img_data: 3D array-like
        The volumetric image data.
    - slice_index: int
        Index of the slice to be visualized.
    - plane: str
        Name of the plane ('axial', 'coronal', 'sagittal').
    """
    # Extract the desired slice based on the plane
    if plane.lower() == "axial":
        slice = img_data[:, :, slice_index]
        slice = ski.transform.rotate(slice, angle=270, resize=True)
        # Crop the slice for axial
        start_row, end_row = 190, 260
        start_column, end_column = 300, 395

    elif plane.lower() == "coronal":
        slice = img_data[:, :, slice_index]
        slice = ski.transform.rotate(slice, angle=-90, resize=True)
        slice_index = np.flipud(slice)
        # Crop the slice for axial
        start_row, end_row = 150, 270
        start_column, end_column = 300, 395

    elif plane.lower() == "sagittal":
        slice = img_data[:, :, slice_index]
        slice = ski.transform.rotate(slice, angle=90, resize=True)
        # Crop the slice for axial
        start_row, end_row = 148, 273
        start_column, end_column = 195, 260

    else:
        raise ValueError("Invalid plane specified. Choose from 'Axial', 'Coronal', or 'Sagittal'.")

    # Crop the slice
    slice_cropped = slice[start_row:end_row, start_column:end_column]

    # Plotting the slice
    plt.figure(figsize=(5, 2))
    plt.imshow(slice_cropped, cmap='gray')
    plt.title(f'{plane} Slice at Index {slice_index}')
    plt.show()

visualize_slice(img_data_flip, 245, 'axial')
visualize_slice(img_data_flip, 147, 'sagittal')#
```
